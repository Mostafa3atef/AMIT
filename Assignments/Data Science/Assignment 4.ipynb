# Write a Pandas program to compare the elements of the two Pandas Series.

```python
df1 = pd.Series([2, 4, 6, 8, 10])
df2 = pd.Series([1, 3, 5, 7, 10])
print(f'{df1}\n{df2}')

```

# Write a Pandas program to change the data type of given a column or a Series.

```python
s1 = pd.Series(['100', '200', 'python', '300.12', '400'])
print(s1)
s2 = pd.to_numeric(s1,errors='coerce')
print(s2)
```

# Write a Pandas program to display the default index and set a column as an Index in a given dataframe.

```python
df = pd.DataFrame({
    'school_code': ['s001','s002','s003','s001','s002','s004'],
    'class': ['V', 'V', 'VI', 'VI', 'V', 'VI'],
    'name': ['Alberto Franco','Gino Mcneill','Ryan Parkes', 'Eesha Hinton', 'Gino Mcneill', 'David Parkes'],
    'date_Of_Birth': ['15/05/2002','17/05/2002','16/02/1999','25/09/1998','11/05/2002','15/09/1997'],
    'weight': [35, 32, 33, 30, 31, 32],
    'address': ['street1', 'street2', 'street3', 'street1', 'street2', 'street4'],
    't_id':['t1', 't2', 't3', 't4', 't5', 't6']})

print('DataFrame with the default index:\n')
print(df)
```

#Write a Pandas program to display the default index and set a column as an Index in a given dataframe and then reset the index.

```python
df1=df.set_index("t_id")
print('DataFrame after setting t_id as new Index:\n')
print(df1)
```

```python
df2=df1.reset_index(inplace=False)
print('DataFrame after resetting the Index to defualt:\n')
print(df2)
```
