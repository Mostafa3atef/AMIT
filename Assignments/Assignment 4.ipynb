### Create a simple class names calculator

### Create a constructor that prints Welcome message

###  Add 2 methods to the class sum & mull

###  The sum method return the sum of 2 arguments x and y

### The mull method return the multiplication of the arguments x and y

###  Take an object from the class

```python
class Calculator:
  def __init__(self):
    print('Welcome...')
  def sum_method(self,x,y):
    return x+y
  def mull_method(self,x,y):
    return x*y

Cal= Calculator()

x=3
y=2
print(Cal.sum_method(x,y))
print(Cal.mull_method(x,y))
```

### Explain in few words why we call the self in methods
#### Answer: to refer to the instance of the class withing its methods. That means, methods of the class will operate on an instance of the class.

### Create a new class names SciCalc with 3 methods , sum , mull , power all of them takes 2 argument x, y

### Sum return the sum of x and y

###  Mull return the multiplication of x and y

```python
class SciCalc:
  def __init__(self):
    print('Welcome...')
  def sum_method(self,x,y):
    return x+y
  def mull_method(self,x,y):
    return x*y
  def power_method(self,x,y):
    return x**y
```

### Inherit from the Calc class , now remove the unneeded code the the SciCalc after inheriting


```python
class SciCalc(Calculator):
  def __init__(self):
    print('Welcome...')
  def power_method(self,x,y):
    return x**y

num1=3
num2=4
new_cal= SciCalc()

print(new_cal.sum_method(num1,num2))
print(new_cal.mull_method(num1,num2))
print(new_cal.power_method(num1,num2))
```

### Explain in few words what happened after inheriting
#### Answer: The new subclass (SciCalc) can inherit the methods and the attributes of the superclass (Calculator)
