### program to count number of vowels using sets in given string


````python
def vowel_count(word):
  vowels= {'a','e','i','o','u'}
  count_number=0
  for i in word:
    if i in vowels:
      count_number+=1

  print(count_number)


str_input = str(input('enter a word: '))
vowel_count(str_input)
````

### Write a Python program to replace the last value of tuples in a list.

### Sample list: [(10, 20, 40), (40, 50, 60), (70, 80, 90)]

### Expected Output: [(10, 20, 100), (40, 50, 100), (70, 80, 100)]



````python
# Method 1
def replace_last_value1(lists1):
  n=[]
  for i in lists1:
    x=i[:-1]+(100,)
    n.append(x)
  print(n)

l1 = [(10, 20, 40), (40, 50, 60), (70, 80, 90)]
replace_last_value1(l1)


# Method 2
def replace_last_value2(lists2):
  print([t[:-1] + (100,) for t in lists2])

l2 = [(10, 20, 40), (40, 50, 60), (70, 80, 90)]
replace_last_value2(l2)
````
